```python
# ***************************************************************************************Taking user input with Tkinter

root = Tk()

variable1 = StringVar(root)
variable1.set("D20ME II / D20MX") # default value


def my_function():

    global whatDevice
    global deviceNames
    global deviceAddresses
    global d20Type
    
 # [:5] limits input to only 5 character no matter how long the input is

    
    input2 = my_entry2.get()
    whatDevice = str(input2)
    
    input3 = my_entry3.get()
    deviceNames = str(input3)
    
    input4 = my_entry4.get()
    deviceAddresses = str(input4)
    
    input5 = variable1.get()
    d20Type = str(input5)
    
    root.destroy()


# used for determining that user input is a digit
def testVal(inStr,acttyp):
    if acttyp == '1': #insert
        if not inStr.isdigit() and inStr > 10000:
            return False
    return True

def testVal2(inStr,acttyp):
    if acttyp == '1': #insert
        if inStr.isdigit(): 
            return False
    return True



my_label = Label(root, text = "What device(s)? (Follow each device by a comma)")
my_label.grid(row = 0, column = 0)
# validates that user input is only letters with testVal2
my_entry2 = Entry(root, validate="key")
my_entry2['validatecommand'] = (my_entry2.register(testVal2),'%P','%d')
my_entry2.grid(row = 0, column = 1)

my_label = Label(root, text = "Device names - 17 Characters Long (Follow each name by a comma)")
my_label.grid(row = 1, column = 0)
# validates that user input is only letters with testVal2
my_entry3 = Entry(root, validate="key")
my_entry3['validatecommand'] = (my_entry3.register(testVal2),'%P','%d')
my_entry3.grid(row = 1, column = 1)

my_label = Label(root, text = "Their Addresses? (Follow each address by a comma)")
my_label.grid(row = 2, column = 0)
# validates that user input is only letters with testVal2
my_entry4 = Entry(root, validate="key")
#my_entry4['validatecommand'] = (my_entry4.register(testVal2),'%P','%d')
my_entry4.grid(row = 2, column = 1)

my_label = Label(root, text = "D20 Type?")
my_label.grid(row = 3, column = 0)
my_entry5 = OptionMenu(root, variable1, "D20M++", "D20ME II / D20MX")
my_entry5.grid(row = 3, column = 1)


# Variables for data collected from the user to create the config
whatDevice = ""
deviceNames = ""
deviceAddresses = ""
d20Type = ""


my_button = Button(root, text = "Submit", command = my_function)
my_button.grid(row = 4, column = 1)

root.wm_state('iconic')

root.title("D20 DCA Generator")

root.mainloop()


# ***************************************************************************************Taking user input with Tkinter
```

```python
# for downloading to any users Downloads folder
import os
def get_download_path():
    """Returns the default downloads path for linux or windows"""
    if os.name == 'nt':
        import winreg
        sub_key = r'SOFTWARE\Microsoft\Windows\CurrentVersion\Explorer\Shell Folders'
        downloads_guid = '{374DE290-123F-4565-9164-39C4925E467B}'
        with winreg.OpenKey(winreg.HKEY_CURRENT_USER, sub_key) as key:
            location = winreg.QueryValueEx(key, downloads_guid)[0]
        return location
    else:
        return os.path.join(os.path.expanduser('~'), 'downloads')
```

```python
# function for adding SATEC points
# for opening and writing to txt file
import requests
downloads = get_download_path()

def addSatecDCA():
    
    #define a path with new file name
    path = str(downloads) + "\AIDESC.txt"
    deviceNamesList = deviceNames.split(", ")
    deviceAddressesList = deviceAddresses.split(", ")
    f1 = open(path, 'w')

    for i in range(len(deviceNamesList)):
        f1.write(deviceNamesList[i] + " PH A VOLTS\n")
        f1.write(deviceNamesList[i] + " PH B VOLTS\n")
        f1.write(deviceNamesList[i] + " PH C VOLTS\n")
        f1.write(deviceNamesList[i] + " PH A AMPS\n")
        f1.write(deviceNamesList[i] + " PH B AMPS\n")
        f1.write(deviceNamesList[i] + " PH C AMPS\n")
        f1.write(deviceNamesList[i] + " PH A KW\n")
        f1.write(deviceNamesList[i] + " PH B KW\n")
        f1.write(deviceNamesList[i] + " PH C KW\n")
        f1.write(deviceNamesList[i] + " PH A KVAR\n")
        f1.write(deviceNamesList[i] + " PH B KVAR\n")
        f1.write(deviceNamesList[i] + " PH C KVAR\n")
        f1.write(deviceNamesList[i] + " PH A KVA\n")
        f1.write(deviceNamesList[i] + " PH B KVA\n")
        f1.write(deviceNamesList[i] + " PH C KVA\n")
        f1.write(deviceNamesList[i] + " PH A PF\n")
        f1.write(deviceNamesList[i] + " PH B PF\n")
        f1.write(deviceNamesList[i] + " PH C PF\n")
        f1.write(deviceNamesList[i] + " TOTAL PF\n")
        f1.write(deviceNamesList[i] + " TOTAL KVAR\n")
        f1.write(deviceNamesList[i] + " TOTAL KVA\n")
        f1.write(deviceNamesList[i] + " NEUT CURRENT\n")
        f1.write(deviceNamesList[i] + " FREQUENCY\n")
        f1.write(deviceNamesList[i] + " ONE WAY PATH DELAY\n")
        f1.write(deviceNamesList[i] + " # SUCCESSFUL TRANS.\n")
        f1.write(deviceNamesList[i] + " # COMM FAILURES\n")
        f1.write(deviceNamesList[i] + " PRESENT CHANNEL\n")
        f1.write(deviceNamesList[i] + " % SUCCESSFUL TRANS.\n")
    f1.close() 

    #define a path with new file name
    path = str(downloads) + "\DIDESC.txt"
    f2 = open(path, 'w')

    f2.write("GLOBAL DEVICES DISABLED\n")
    f2.write("GLOBAL POLLING DISABLED ALL DEV\n")
    f2.write("GLOBAL DISABLE UNSOL RESP ALL DEV\n")

    for i in range(len(deviceNamesList)):
        f2.write(deviceNamesList[i] + " ENABLED/DISABLED DEVICE\n")
        f2.write(deviceNamesList[i] + " POLLING ENAB/DISAB\n")
        f2.write(deviceNamesList[i] + " UNSOL RESP ENAB/DISAB\n")
        f2.write(deviceNamesList[i] + " POLL FOR DATA\n")
        f2.write(deviceNamesList[i] + " REMOTE RESTART\n")
        f2.write(deviceNamesList[i] + " TIME SYNC REMOTE\n")
        f2.write(deviceNamesList[i] + " PROPAGATION DELAY MEASURE\n")
        f2.write(deviceNamesList[i] + " PRI COMM CHANNEL STATUS\n")
        f2.write(deviceNamesList[i] + " SEC COMM CHANNEL STATUS\n")
        f2.write(deviceNamesList[i] + " PRI CHANNEL IN USE\n")
        f2.write(deviceNamesList[i] + " SEC CHANNEL IN USE\n")
        f2.write(deviceNamesList[i] + " IIN RESTART BIT STATE\n")
        f2.write(deviceNamesList[i] + " IIN TROUBLE BIT STATE\n")
        if d20Type == "D20ME II / D20MX":
            f2.write(deviceNamesList[i] + " IIN LOCAL BIT STATE\n")
            f2.write(deviceNamesList[i] + " IIN BUFFER OVERFLOW\n")
            f2.write(deviceNamesList[i] + " IIN CONFIG CORRUPT\n")

    f2.close() 

    #define a path with new file name
    path = str(downloads) + "\DODESC.txt"
    f3 = open(path, 'w')

    f3.write("GLOBAL DISABLE ALL DEV\n")
    f3.write("GLOBAL DISABLE POLLS TO ALL DEV\n")
    f3.write("GLOBAL DISABLE UNSOL RESP\n")
    f3.write("GLOBAL POLL ALL DEV FOR CLSS DATA\n")
    f3.write("GLOBAL POLL ALL DEV FOR INTG DATA\n")
    f3.write("GLOBAL RESTART ALL DEV\n")
    f3.write("GLOBAL TIMESYNC ALL DEV\n")
    f3.write("GLOBAL CLEAR CHANNEL\n")
    f3.write("GLOBAL CLEAR COMM STATISTIC\n")

    for i in range(len(deviceNamesList)):
        f3.write(deviceNamesList[i] + " DEVICE ENABLED\n")
        f3.write(deviceNamesList[i] + " SCHEDULED POLLS ENABLED\n")
        f3.write(deviceNamesList[i] + " UNSOLICTED RESP ENABLED\n")
        f3.write(deviceNamesList[i] + " POLL DEV FOR CLASS DATA\n")
        f3.write(deviceNamesList[i] + " POLL DEV FOR INTEG DATA\n")
        f3.write(deviceNamesList[i] + " DEVICE RESTART\n")
        f3.write(deviceNamesList[i] + " TIMESYNC IN PROGRESS\n")
        f3.write(deviceNamesList[i] + " PROPAGATION DELAY IN PROG\n")
        f3.write(deviceNamesList[i] + " FORCE NEXT COM CHANNEL\n")
        f3.write(deviceNamesList[i] + " CLEAR COMM STATISTIC\n")

    f3.close() 

    #define a path with new file name
    path = str(downloads) + "\DVDESC.txt"
    f4 = open(path, 'w')
    
    for i in range(len(deviceNamesList)):
        f4.write(deviceNamesList[i] + " SATEC METER (Addr. " + deviceAddressesList[i] + " )\n" )
    
```
